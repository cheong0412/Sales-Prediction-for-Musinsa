# 통계적 유의미성 판단하는 방법
통계적 유의성을 판단할 때는 P-value와 t-통계량을 함께 고려해야 합니다.
- P-value가 작고 t-통계량이 크면 해당 독립 변수는 통계적으로 유의미하다고 판단할 수 있습니다.

---

#### **cf) Prob.F-Statistic**
- 회귀모형에 대한 **통계적 유의미성** 검정.
- 이 값이 ***0.05 이하라면 모집단에서도 의미가 있다고 볼 수 있음.***

#### **1. R-squared (R²)**
- 종속 변수의 변동 중 모델이 설명하는 비율을 나타내는 지표입니다.
- ***0에서 1 사이의 값을 가지며, 1에 가까울수록 모델이 데이터를 잘 설명한다고 할 수 있습니다.***
- 하지만 과적합을 방지하기 위해 다른 평가 지표와 함께 고려하는 것이 좋습니다.

#### 2. Adj. R-squared (Adjusted R²)
- R-squared를 보완한 지표로, 독립 변수의 수를 고려한 조정된 R-squared입니다.
- 모델에 추가된 독립 변수가 설명력을 향상시키지 못하는 경우, Adjusted R-squared는 낮아집니다.

#### **3. 회귀 계수 (Coefficients)**
- 각 독립 변수의 회귀 계수를 확인하여 해당 독립 변수가 종속 변수에 미치는 영향을 이해할 수 있습니다.
- ***계수의 부호와 크기를 고려하며, 통계적으로 유의미한지 여부를 P-value와 t-통계량으로 확인합니다.***
#### 해당 독립 변수의 회귀 계수가 양수이면, 해당 독립 변수가 종속 변수에 양의 영향을 미친다고 해석할 수 있습니다. 음수이면 음의 영향을 미친다고 해석할 수 있습니다.

#### 4. 표준 오차 (Standard Error)
- ***회귀 계수의 추정의 불확실성을 나타내는 지표입니다. 작을수록 좋습니다.***
#### 표준 오차가 작을수록 해당 계수의 추정이 더 정확하다고 할 수 있습니다.

#### 5. P-value (P>|t|)
- 귀무 가설이 "해당 독립 변수의 회귀 계수는 0이다"일 때의 확률값입니다.
- P-value가 작을수록 해당 독립 변수의 회귀 계수가 통계적으로 유의미하다고 할 수 있습니다.
#### P-value가 작을수록 해당 계수가 통계적으로 유의미하다고 판단됩니다.

#### 6. F-statistic 및 Prob (F-statistic)
- F-statistic은 모든 독립 변수의 유의성을 동시에 검정하는 통계량입니다.
- Prob (F-statistic)은 F-statistic의 유의성을 나타내는 P-value입니다.

#### 7. 잔차 진단 (Residual Diagnostics)
- 잔차 분석을 통해 모델이 가정을 만족하는지 확인합니다.
- 잔차의 정규성, 등분산성, 독립성 등을 검토합니다.

#### 8. Durbin-Watson 통계량
- 잔차 간의 자기 상관을 검사하는 통계량으로, 2에 가까울수록 이상적입니다.
#### t-통계량 (t):
- 회귀 계수를 표준 오차로 나눈 값으로, 회귀 계수가 0으로부터 얼마나 떨어져 있는지를 나타냅니다.
- t-통계량이 크면 해당 독립 변수의 회귀 계수가 0이 아닌 것으로 간주되며, 작으면 회귀 계수가 통계적으로 유의하지 않다고 판단될 수 있습니다.
#### t-통계량이 크면 해당 계수가 통계적으로 유의미하다고 판단됩니다.
